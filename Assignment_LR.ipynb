```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
df= pd.read_csv('/content/data.csv')
df.head()
```

task - 1

```python
x = df['weight']
y = df['price']

x_mean = np.mean(x)
y_mean = np.mean(y)

m = np.sum(  (x-x_mean)*(y-y_mean) ) / np.sum(  (x-x_mean)**2 )
c = y_mean - m*x_mean

print(m)
print(c)
```

```python
# for vegetable weight is 6.

y= m*6+c
print(y)
```

task - 2

```python
reg.fit(df[['weight']],df['price'])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
df.head()
```

```python
df['Predicted_price'] = reg.predict(df[['weight']])
df.head()
```

```python
df['residuals'] = df['price'] - df['Predicted_price']
df.head()
```

```python
plt.scatter(df['weight'],df['price'],label = "Row Data")
plt.scatter(df['weight'],df['Predicted_price'], label="Predicted Data")
plt.xlabel('weight')
plt.ylabel('price')
plt.plot(df['weight'],df['Predicted_price'],color='red',label = 'Best FIt line')
plt.legend()
plt.show()
```

task - 3

```python
MSE = mean_squared_error(df['price'],df['Predicted_price'])
print("MSE = ",MSE)

MAE = mean_absolute_error(df['price'],df['Predicted_price'])
print("MAE = ",MAE)
```
