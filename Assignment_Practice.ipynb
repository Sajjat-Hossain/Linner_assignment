```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
reg = linear_model.LinearRegression()
```

```python
df = pd.read_csv('/content/Dhaka Rent.csv',sep=';')
df.head(5)
```

```python
x = df['area']
y = df['rent']

mean_x = np.mean(x)
mean_y = np.mean(y)

M = np.sum( (x-mean_x)*(y-mean_y) ) / np.sum( (x-mean_x)**2 )

c = mean_y - M*mean_x
print('M = ',M)
print('C = ',c)


# For 2000 value
y = M*2000 + c;
print(y)
```

```python
plt.scatter(df.area,df.rent)
plt.xlabel('Area')
plt.ylabel('Rent')
```

```python
reg.fit(df[['area']],df.rent)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2000]])
```

```python
df['Predicted Rent'] = reg.predict(df[['area']])
df.head(5)
```

```python
df['Error'] = df['rent'] - df['Predicted Rent']
df.head(5)
```

```python
mse = mean_squared_error(df.rent,df['Predicted Rent'])
mse
```
